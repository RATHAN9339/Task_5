```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("data.csv")
df.head(5)
```

```python
df.tail(5)
```

```python
df.shape
```

```python
#Duplicate rows
print("No of Duplicate rows in the Dataset:",df.duplicated().sum())
```

```python
df.count()
```

```python
#Drop the Duplicate rows
df=df.drop_duplicates()
df.count()
```

```python
print("No of Duplicate rows in the Dataset:",df.duplicated().sum())
```

```python
df.isnull().sum()
```

```python
#Drop the null values
df.dropna(inplace=True)
```

```python
print("No of Null values in the Dataset:",df.isnull().sum().sum())
```

```python
df.info()
```

```python
df.describe()
```

Detecting Outliers

```python
sns.boxplot(x=df['Popularity'])
```

```python
sns.boxplot(x=df['Engine HP'])
```

Bar Plot

```python
df.Make.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of cars by make")
plt.ylabel('Number of cars')
plt.xlabel('Make');
```

Histogram

```python
plt.hist(df['Driven_Wheels'],color="coral")
plt.xlabel("Driven_Wheels")
plt.ylabel("Frequency")
plt.show()
```

Scatter Plot

```python
plt.scatter(df["Engine HP"],df['MSRP'],color="purple")
plt.xlabel("Engine HP")
plt.ylabel("MSRP")
plt.figure(figsize=(10,6))
plt.show();
```
